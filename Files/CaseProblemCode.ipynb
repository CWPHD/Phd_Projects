```python
#Clusters
from kmodes.kmodes import KModes
import matplotlib.pyplot as plt

# Load the main dataset
file_path = '/Users/cwaldner/Sandbox/TIM_8501/Week8/cps.csv'
df = pd.read_csv(file_path)

# Columns for clustering
columns_for_clustering = ['PRPERTYP', 'HETENURE', 'HEHOUSUT', 'HEFAMINC', 'HRNUMHOU', 'HRHTYPE', 'HUBUS', 'GEREG', 'PEHRFTPT', 'PRMJOCGR', 'HESS3']

# Range of clusters
clusters_range = range(1, 11)  

# Store the cost values for each cluster
costs = []

# Calculate cost for each number of clusters
for num_clusters in clusters_range:
    kmodes = KModes(n_clusters=num_clusters, init='Cao', verbose=0)
    clusters = kmodes.fit_predict(df[columns_for_clustering])
    costs.append(kmodes.cost_)

# Plot the elbow graph
plt.plot(clusters_range, costs, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Cost')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.show()
```
